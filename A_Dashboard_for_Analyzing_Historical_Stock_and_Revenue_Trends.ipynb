```python
import yfinance as yf

tesla_data = yf.download('TSLA', start='2022-01-01', end='2023-12-31')

tesla_data.reset_index(inplace=True)

tesla_data.to_csv('tesla_data.csv', index=False)

print(tesla_data.head())
```

```python
import yfinance as yf

tesla = yf.Ticker("TSLA")

financials_annual = tesla.financials
financials_quarterly = tesla.quarterly_financials

print("Annual Financials (most recent):")
print(financials_annual.loc['Total Revenue'].tail())

print("\nQuarterly Financials (most recent):")
print(financials_quarterly.loc['Total Revenue'].tail())
```

```python
import yfinance as yf

# Fetch historical data for GameStop
gme_data = yf.download('GME', start='2022-01-01', end='2022-12-31')

# Reset the index to change the Date to column
gme_data.reset_index(inplace=True)

# Save the dataframe as CSV file
gme_data.to_csv('gme_data.csv', index=False)

# Display first five rows
print(gme_data.head())
```

```python
import yfinance as yf

gme = yf.Ticker("GME")

financials_quarterly = gme.quarterly_financials

print("Quarterly Financials (Revenue, last 5 entries):")
if 'Total Revenue' in financials_quarterly.index:
    print(financials_quarterly.loc['Total Revenue'].tail())
else:
    print("Revenue data not available.")
```

```python
import yfinance as yf
import matplotlib.pyplot as pl
tesla_data = yf.download('TSLA', start='2022-01-01', end='2022-12-31')
plt.figure(figsize=(10, 5))
plt.plot(tesla_data['Close'], label='Tesla Close Price')
plt.title('Tesla Stock Close Price (2022)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
gme_data = yf.download('GME', start='2022-01-01', end='2022-12-31')
plt.figure(figsize=(10, 5))
plt.plot(gme_data['Close'], label='GameStop Close Price')
plt.title('GameStop Stock Close Price (2022)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
